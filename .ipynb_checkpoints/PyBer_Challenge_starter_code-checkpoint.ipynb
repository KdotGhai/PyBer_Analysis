# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import requests
import json
import statistics
import numpy as np

# File to Load (Remember to change these)
city_data_to_load ="Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_by_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_by_city_type
```

```python
# 2. Get the total drivers for each city type
total_drivers_by_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_by_city_type
```

```python
#  3. Get the total amount of fares for each city type
total_fares_by_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_by_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_ride = (total_fares_by_city_type / total_rides_by_city_type)
avg_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_drivers = (total_fares_by_city_type / total_drivers_by_city_type)
avg_fare_per_drivers
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({ 
    "Total Rides": total_rides_by_city_type,
    "Total Drivers": total_drivers_by_city_type,
    "Total Fares" : total_fares_by_city_type,
    "Average Fare per Ride":avg_fare_per_ride,
    "Average Fare per river": avg_fare_per_drivers})

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:.1f}".format)

pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:.1f}".format)

pyber_summary_df["Total Fares"] = pyber_summary_df[ "Total Fares"].map("${:,.2f}".format)

pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)

pyber_summary_df["Average Fare per river"] = pyber_summary_df["Average Fare per river"].map("${:.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame 
pyber_data_df.head()
```

```python
    # combine "city type(city)" with "date" using groupby()
    # while using groupby() take the sum of "fare"
    
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

city_fares_df = pyber_data_df.groupby([ "date","type"]).sum()[["fare"]]
city_fares_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
city_fares_df = city_fares_df.reset_index()
city_fares_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

city_fares_df_pivot = city_fares_df.pivot(index = 'date', columns = 'type', values = 'fare')
city_fares_df_pivot.head()
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
city_fares_pivot_new = city_fares_df_pivot.loc[ '2019-01-01':'2019-04-29']
city_fares_pivot_new
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

city_fares_pivot_new.index = pd.to_datetime(city_fares_pivot_new.index)
city_fares_pivot_new
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
city_fares_pivot_new.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
city_fares_pivot_weeks =  city_fares_pivot_new.resample("W").sum()
city_fares_pivot_weeks 
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 


# Import the style from Matplotlib.
from matplotlib import style

# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
    # This style.use() lets me preload colors and labels needed for chart beforehand
    # Saves user the trouble of having to place all the parameters into plt.plot() or where "graph_dates = " is




graph_dates =  city_fares_pivot_weeks.plot(kind = 'line', figsize=(14, 9))

graph_dates.set_title(' Total Fares by City Type(2019)',fontsize=20)
graph_dates.set_ylabel('Fares ($USD)',fontsize=14)
graph_dates.set_xlabel("Date",fontsize=14)

#Save image of chart
plt.savefig("analysis/PyBer_fare_summary.png", bbox_inches='tight', pad_inches=0.5)


```
