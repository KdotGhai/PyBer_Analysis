```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)



plt.show()
```

```python
###### Create the plot with ax.plt()  ##This is creating a Tuple(data that cannot be changed and uses much less memory)

## Here we're asking to produce a figure just like above but it has no parameter for X and Y axis
fig, ax = plt.subplots()

###ax.plt() will import the data that will be assigned to X and Y axis
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plot()   ###Alternative to using subplot,fig, and ax
## Set Size  of Figure with figsize inside of plt.figure(figsize(width,height))
fig = plt.figure(figsize=(12,12))  #will generate blank chart/figure
ax = fig.add_subplot() #is appending data needed to fill in chart/figure(currently empty)



#plt.show()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()  ###Another Alternative for quickly making chart with parameters
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

#### Set Size of plot chart
plt.rcParams['figure.figsize'] = [12, 12]

# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()

plt.show()
```

```python
### Amending Chart above to have grid, colors = blue, markers ="*"


# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

plt.rcParams['figure.figsize'] = [20, 20]

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()

```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.scatter(x_axis, y_axis )
#Alternative but not desired method: plt.plot(x_axis, y_axis, 'o')
# the 'o' is simply an argument to remove the lines and only show plot points
```

```python
###### Skill Excercise
#Change the color of the markers to red.

#Add a legend for the city of Chicago.

#Add a title and axes labels.

#Switch the axis so that the Fare($) data is on the x-axis.

#Add a limit for the x-axis data.

#Invert the y-axis so that January is at the top.
```

```python
#create graph
plt.scatter(y_axis, x_axis,  color = "red",label= "Chicago" , s = 50)
#Set labels
plt.title("PyBer by Month")
plt.ylabel("Months")
plt.xlabel("Fare($)")
#inverse data for y-axis
plt.gca().invert_yaxis()
#show legend
plt.legend()
```

```python
#### Manipulate size of data points based on axis data

plt.scatter(x_axis, y_axis, s=y_axis)

#The data points change in size but its not clean or presentable
```

```python
### We're going to increase the size of y values (Still a scatter plot but now is know as Bubble Chart)
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

### Then display again the data with the new  y_axis_larger which is simply 3 times of y_axis original values
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
### Instead of using For/while loops python is capable of recognizing patterns thus we can use list comprehension

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])

# We produced the same data results as above
```

```python
### Now to create (Obejct orientated interface) or tuple() based scatter plot

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = y_axis)
```

```python
#### skill excercise

#Change the color of the markers to sky blue.

#Change the size of the markers to 5 times each data point.

#Make the color 20% transparent.

#Add a black edge color to the circles.

#Make the linewidth of the circles 2 points.

#Add a legend for the city of Boston.

#Add a title.

#Switch the axis so that the Fare($) data is on the x-axis.

#Add a limit for the x-axis data.

#Invert the y-axis so that January is at the top.
```

```python
# (Still a scatter plot but now is know as Bubble Chart)
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis,  color = 'xkcd:sky blue' ,  alpha= .8 , edgecolors= "black", linewidths= 2
           , label= "Boston",s = [i * 5 for i in y_axis])

#set labels
plt.title("PyBer by Month", weight = 'bold', fontsize=25)

plt.ylabel("Months", weight = 'bold', fontsize=20) #ADjust fontsize and bold xlabel
plt.xlabel("Fare($)", weight = 'bold', fontsize=20)


#inverse data for y-axis
plt.gca().invert_yaxis()
#set xlim to 50
ax.set_xlim(0, 50)
#show legend
ax.legend()
```

```python
### Error Bars to line Chart
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
### let's add error bars and a capsize using the object-oriented approach.

fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
## Add Error Bars to a Bar Chart
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
### let's add error bars and a capsize using the object-oriented approach.

fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
## Change the Major Ticks
```

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()

plt.show()
```

```python
### create the same graph using the object-oriented approach

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
### Add Minor Ticks

from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(10, 10))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
