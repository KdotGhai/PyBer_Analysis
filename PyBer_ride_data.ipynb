```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
import json
import statistics

# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")

city_data_df = pd.read_csv("Resources/city_data.csv")

ride_data_df = pd.read_csv("Resources/ride_data.csv")

pyber_ride_df
```

```python
print(pyber_ride_df.dtypes) 
```

```python
## Creating line chart

pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.

# Here we  are making an array to store x values
x_axis = np.arange(len(pyber_ride_df))

tick_locations = [value for value in x_axis]

# Plot the data.

pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")

plt.rcParams['figure.figsize'] = [10, 10]

plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
### Creating Bar Chart
# Method #1: Bar Chart
# We are chaining .bar() to .plot thus setting the chart type for the dataframe
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Method 2: Bar Chart
# here we use "kind=" to state what chart type we want

pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.show()
```

```python
# Skill Drill

# Add error bars.
# Add caps to the error bars.
# Change the color of the bars to sky blue.
# Rotate the labels on the x-axis to horizontal.
# Set the y-axis increment to every $5.
```

```python
#Error bars

# get standard deviation of data first
stdev = statistics.stdev(pyber_ride_df['Avg. Fare ($USD)'])
#stdev



pyber_ride_df.plot.bar(x = 'Month', y='Avg. Fare ($USD)' , yerr=stdev, capsize=3, color= "skyblue")
plt.yticks(np.arange(0, 51, step=5.0))

plt.show()
```
